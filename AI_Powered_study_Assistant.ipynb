Import Module from Gemini Package

```python
from google import genai
from google.colab import userdata
```

Use Client class to create an instance(for interacting with Gemini API)

```python
Client = genai.Client(api_key =userdata.get('GEMINI_API_KEY'))
print(Client)
```

Use Client Object for communicating to the Gemini LLM

```python
def study_assistant(user_prompt):
    response = Client.models.generate_content(
        model="gemini-3.5-flash",
        contents=user_prompt
    )
    return response
```

```python
output = study_assistant("Explain about the LLM Applications")
print(output.text)
```
